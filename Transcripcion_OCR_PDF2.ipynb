```python
# ðŸ”¹ Montar Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# ðŸ”¹ Instalar Tesseract OCR y PyMuPDF
!apt-get install -y tesseract-ocr tesseract-ocr-spa
!pip install pytesseract pymupdf

# ðŸ”¹ Importar LibrerÃ­as
import pytesseract
import fitz  # PyMuPDF
from PIL import Image
import os
import cv2
import numpy as np

# ðŸ”¹ Configurar Tesseract en espaÃ±ol
os.environ["TESSDATA_PREFIX"] = "/usr/share/tesseract-ocr/4.00/tessdata/"

# ðŸ”¹ Ruta del PDF en Google Drive
pdf_path = "/content/drive/My Drive/Mariano/Leon.pdf"

# ðŸ”¹ Abrir el PDF
doc = fitz.open(pdf_path)

# ðŸ”¹ Variable para guardar la transcripciÃ³n
output_text = ""

# ðŸ”¹ Procesar cada pÃ¡gina del PDF
for i in range(len(doc)):
    print(f"ðŸ“„ Procesando pÃ¡gina {i+1}/{len(doc)}...")

    # ðŸ”¹ Extraer imagen de la pÃ¡gina
    pix = doc[i].get_pixmap(dpi=300)  # Puedes aumentar DPI para mejorar la calidad
    img_path = f"/content/temp_page_{i+1}.png"
    pix.save(img_path)

    # ðŸ”¹ Mejorar la imagen (Escala de grises y binarizaciÃ³n)
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)

    # ðŸ”¹ Guardar imagen mejorada
    enhanced_img_path = f"/content/enhanced_page_{i+1}.png"
    cv2.imwrite(enhanced_img_path, thresh)

    # ðŸ”¹ Aplicar OCR con Tesseract
    text = pytesseract.image_to_string(Image.open(enhanced_img_path), lang="spa")

    # ðŸ”¹ Agregar texto al resultado final
    output_text += f"\n\n--- PÃ¡gina {i+1} ---\n{text}"

    # ðŸ”¹ Eliminar archivos temporales
    os.remove(img_path)
    os.remove(enhanced_img_path)

# ðŸ”¹ Guardar la transcripciÃ³n en Google Drive
output_text_path = "/content/drive/My Drive/Mariano/Leon.pdf_Transcripcion_Tesseract.txt"
with open(output_text_path, "w", encoding="utf-8") as text_file:
    text_file.write(output_text)

print(f"âœ… TranscripciÃ³n completada con Tesseract OCR. Revisa en tu Drive: {output_text_path}")
```
